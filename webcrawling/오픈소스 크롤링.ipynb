```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests as req
import time
```

```python
ossurl = 'https://olive.kakao.com/component/list?page='
oliveurl = 'https://olive.kakao.com'
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.implicitly_wait(10)
driver.get(ossurl)
```

새창에서 카카오 로그인

```python
ossdata = []
ossdata.append(['name','url','license','copyright'])
```

```python
dependata = []
dependata.append(['name','groupid','artifactid','version','license','filetype','packagemanager'])
```

```python
nodata = 0
for i in range(333):
    print(i/333 * 100)
    url = ossurl+str(i)
    driver.get(url)
    time.sleep(1.0)
    
    soup = bs(driver.page_source,'html.parser')
    datalen = len(soup.select('table > tbody > tr > td > span > component-tag + a'))
    for detail in soup.select('table > tbody > tr > td > span > component-tag + a'):
        detailurl = oliveurl+detail['href']
        driver.get(detailurl)
        #driver.get('https://olive.kakao.com/component/detail?id=11845')
        time.sleep(1.5)
        soup = bs(driver.page_source,'html.parser')
        data = soup.select('div.ant-descriptions-view > table > tbody > tr + tr > td')
        name = data[0].text
        url = data[1].text
        license = data[2].text.strip(' ')
        copyright = data[3].text
        ossdata.append([name,url,license,copyright])

        
        data = soup.select('div.ant-table-content > table > tbody > tr')
        if(data[0].text == ' 데이터 없음 '):
            nodata += 1
            continue
        for tr in data :
            if('data-index' in tr.attrs):
                soup2 = bs(str(tr),'html.parser')
                td = soup2.select('td')
                depen = []
                depen.append(name)
                for tdidx in [2,3,4,5,6,7] :
                    soup3 = bs(str(td[tdidx]),'html.parser')
                    depen.append(soup3.select('span + span ')[0].text.strip(' '))
                dependata.append(depen)

```

```python
df = pd.DataFrame(ossdata)
df.to_csv('ossdata.csv')
```

```python
df = pd.DataFrame(dependata)
df.to_csv('dependata.csv')
```
